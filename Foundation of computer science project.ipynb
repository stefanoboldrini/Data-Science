```python
# Point n.1: Convert the app sizes to a number  

# Con il seguente definisco la ricerca effettuata tramite regex, definisco i valori M e k e applico la funzione 
human_to_int_re = re.compile('(?P<amount>\d+\.{0,1}\d*)(?P<unit>\w{0,1})')

def unita(unit):
    
    if unit == 'M':
        return 1000000
    if unit == 'k':
        return 1000
    return 1

def to_numeric(elem):
    m = human_to_int_re.search(elem)
    if m is None:
        return None
    unit = m.group('unit')
    mult = unita(unit)
    amount = float(m.group('amount'))
    return int(amount * mult)

```

```python
#Applico la funzione "to numeric" alla colonna Size
googleplay['Size'] = googleplay['Size'].apply(to_numeric)
```

```python
#Modifico i valori con NA in -1 (ai fini dell'elaborato, non essendo più necessaria la colonna Size, trasformo i valori NA nel valore -1, in modo tale da poter applicare la trasformazione)
googleplay['Size'] = googleplay['Size'].fillna(-1)
```

```python
# Point n. 11. For each app, compute the approximate income, 
##obtain as a product of number of installs and price.
#Creo la funzione che converte in numero come analogamente fatto nel precedente punto n.1

price_to_number = re.compile('\D*(?P<amount>\d+\.{0,1}\d*)')

def convertitore(price):
    valore = price_to_number.search(price)
    if valore:
        return float(valore.group('amount'))
    else:
        return None
```

```python
# Applico la funzione alla colonna Price, sostituendola con quella precedente
googleplay['Price']=googleplay.Price.apply(convertitore) 
```
