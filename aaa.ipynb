```python
### Point n.1: Convert the app sizes to a number  

## Con il seguente definisco la ricerca effettuata tramite regex, definisco i valori M e k
## e applico la funzione 
human_to_int_re = re.compile('(?P<amount>\d+\.{0,1}\d*)(?P<unit>\w{0,1})')

def unita(unit):
    
    if unit == 'M':
        return 1000000
    if unit == 'k':
        return 1000
    return 1

def to_numeric(elem):
    m = human_to_int_re.search(elem)
    if m is None:
        return None
    unit = m.group('unit')
    mult = unita(unit)
    amount = float(m.group('amount'))
    return int(amount * mult)

```

```python
googleplay['Size'] = googleplay['Size'].apply(to_numeric)
```

```python
googleplay['Size'] = googleplay['Size'].fillna(-1)
```
